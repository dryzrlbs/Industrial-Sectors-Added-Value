```python
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource, HoverTool, LinearColorMapper

def color_gen(x):
    if x == "imalat":
        return "red"
    elif x== "tarım":
        return "green"
    elif x == "ins_en_ç":
        return "blue"
    elif x=="hizm":
        return "purple"
    elif x== "maden":
        return "black"


x = df4[df4["yıl"]==2018]["yabancı_g"].values
y = df4[df4["yıl"]==2018]["yabancı_ç"].values
colors = [color_gen(x) for x in df4[df4["yıl"]==2018]["sınıf2"].tolist()]
desc=df4[df4["yıl"]==2018].index.tolist()

source=ColumnDataSource(data=dict(x=x, y=y, desc=desc, color=colors))

hover = HoverTool(tooltips=[("index","$index"),("(x,y)", "(@x, @y)"),("desc","@desc")])

p = figure(plot_width=800, plot_height=800, tools=[hover,"pan","wheel_zoom","box_zoom","reset","save"], x_range=(-0.1,1.1), y_range=(-0.1,1.1))

p.xaxis.axis_label="yabancı girdi"
p.yaxis.axis_label="yabancı çıktı"

p.circle("x","y", color="color", size=10, fill_alpha=0.2, source=source)
p.line([x.mean(), x.mean()],[-0.1,1.1], alpha=1, color="grey")
p.line([-0.1,1.1],[y.mean(), y.mean()], alpha=1, color="grey")

show(p)
```
